```python
# Load the dataset
file_path = "/content/UsedCars_Data.csv"
df = pd.read_csv(file_path)

# Split the 'title' column
df[['Brand', 'Model']] = df['title'].str.split(' ', n=2, expand=True).iloc[:, 1:]
```

```python
def clean_price(price):
    if isinstance(price, str):
        price = price.replace("₹", "").replace(",", "").strip()  # Remove ₹ and commas

        # Check for the presence of known terms (Lakh, Crore, etc.)
        if "Lakh" in price:
            try:
                return float(price.replace("Lakh", "").strip()) * 1e5  # Convert Lakh to actual number
            except ValueError:
                return None  # In case of invalid conversion
        elif "Crore" in price:
            try:
                return float(price.replace("Crore", "").strip()) * 1e7  # Convert Crore to actual number
            except ValueError:
                return None  # In case of invalid conversion
        elif "Thousand" in price:
            try:
                return float(price.replace("Thousand", "").strip()) * 1e3  # Convert Thousand to actual number
            except ValueError:
                return None  # In case of invalid conversion
        else:
            # Handle values like ₹100 or ₹15.50 directly
            try:
                return float(price)
            except ValueError:
                return None  # If it's not convertible, return None

    return None  # Handle missing or invalid values

# Apply the function to the price column
df["price"] = df["price"].apply(clean_price)

# Check the first few rows of the updated price column
print(df["price"].head())
```

```python
# Split the 'title' column and extract Brand and Model
df[['Brand', 'Model']] = df['title'].str.split(' ', n=2, expand=True).iloc[:, 1:]

# Drop the 'title' column
df.drop(columns=['title'], inplace=True)

# Reorder columns to place 'Brand' and 'Model' at the beginning
new_column_order = ['Brand', 'Model'] + [col for col in df.columns if col not in ['Brand', 'Model']]
df = df[new_column_order]
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error


# Initialize the RandomForestRegressor model
model1 = RandomForestRegressor(n_estimators=100,max_depth=5).fit(x_train, y_train)

# Predict the price on the test set
y_pred = model1.predict(x_test)

# Evaluate the model using mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R^2 score
mae = mean_absolute_error(y_test,y_pred)
# Print the evaluation metrics
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
print(f'Mean Absolute Error: {mae}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
model2 = LinearRegression().fit(x_train, y_train)
y_pred = model2.predict(x_test)
# Evaluate the model using mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R^2 score
mae = mean_absolute_error(y_test,y_pred)
# Print the evaluation metrics
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
print(f'Mean Absolute Error: {mae}')
```

```python
import xgboost as xgb
model3 = xgb.XGBRegressor().fit(x_train, y_train)
y_pred = model3.predict(x_test)
# Evaluate the model using mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R^2 score
mae = mean_absolute_error(y_test,y_pred)
# Print the evaluation metrics
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
print(f'Mean Absolute Error: {mae}')
```
